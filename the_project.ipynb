```python
import re
```

```python
data = re.sub(r'[^\w\s]', '', data) #удаляем все знаки препинания
```

```python
data = data.lower() #всё в нижний регистр
pattern = r'[0-9]'
data = re.sub(pattern, '', data) #удаление всех цифр
```

```python
from nltk.tokenize import word_tokenize

tokens_nltk = word_tokenize(data)
tokens_nltk = list(filter(lambda x: not re.search("[a-zA-Z]", x), tokens_nltk))
```

```python
for word in tokens_nltk:
  s=word[0]
  if s=='_':
    tokens_nltk.remove(word)
```

```python
def count_words(list):
    dicti = {}
    for word in list:
        if word in dicti:
            dicti[word] += 1
        else:
            dicti[word] = 1
    return dicti
```

```python
a=count_words(tokens_nltk)
```

запишем словарь в файл

```python
def write_dictionary_to_file(dictionary, file_name):
    with open(file_name, 'w') as file:
        for key, value in dictionary.items():
            file.write(f'{key}: {value}\n')
```
